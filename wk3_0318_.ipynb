# Perceptron
- AND GATE : 
- OR GATE : 하나라도 '1'이 있으면 1
- EXCLUSIVE OR GATE (XOR GATE) : 중간 layer를 여러 개 만드는 방법 밖에 없다. => "Multi perceptron"

```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```
# AND GATE
x = np.array([[1,1],[1,0],[0,1],[0,0]])
y = np.array([[1],[0],[0],[0]])

w = tf.random.normal([2],0,1) # 평균:0, 표준편차:1 (gaussian distribution)에서 랜덤 값 2개 뽑기 -> [2]
b = tf.random.normal([1],0,1)
b_x = 1

for i in range(2000):
  error_sum = 0
  for j in range(4):
    a = np.sum(x[j]*w) + b_x*b
    output = 1/(1+np.exp(-a)) #sigmoid 함수 f(a)
    error = y[j][0] - output
    
    w = w + x[j] * 0.1 * error # 0.1:학습값?
    b = b + b_x * 0.1 * error
    
    error_sum = error_sum + error

  if i % 200 == 199: # 총 2000번 돌리는데 에러 확인을 위해 200번 마다 error print (나머지=199)
    print(i, error_sum, w.numpy(), b.numpy())

# w가 줄어드는 것이 error를 줄일 수 있다.
```

```
#확인용
for i  in range(4):
  a = np.sum(x[i]*w)+b_x*b
  print('X:', x[i], 'Y:', y[i], 'predicted:', 1/(1+np.exp(-a)))
```

```
# OR GATE
x = np.array([[1,1],[1,0],[0,1],[0,0]])
y = np.array([[1],[1],[1],[0]])

w = tf.random.normal([2],0,1) # 평균:0, 표준편차:1 (gaussian distribution)에서 랜덤 값 2개 뽑기 -> [2]
b = tf.random.normal([1],0,1)
b_x = 1

for i in range(2000):
  error_sum = 0
  for j in range(4):
    a = np.sum(x[j]*w) + b_x*b
    output = 1/(1+np.exp(-a)) #sigmoid 함수 f(a)
    error = y[j][0] - output
    
    w = w + x[j] * 0.1 * error # 0.1:학습값?
    b = b + b_x * 0.1 * error
    
    error_sum = error_sum + error

  if i % 200 == 199: # 총 2000번 돌리는데 에러 확인을 위해 200번 마다 error print (나머지=199)
    print(i, error_sum, w.numpy(), b.numpy())
```

```
#확인용
for i  in range(4):
  a = np.sum(x[i]*w)+b_x*b
  print('X:', x[i], 'Y:', y[i], 'predicted:', 1/(1+np.exp(-a)))
```

```
# XOR GATE
x = np.array([[1,1],[1,0],[0,1],[0,0]])
y = np.array([[0],[1],[1],[0]])

w = tf.random.normal([2],0,1) # 평균:0, 표준편차:1 (gaussian distribution)에서 랜덤 값 2개 뽑기 -> [2]
b = tf.random.normal([1],0,1)
b_x = 1

for i in range(2000):
  error_sum = 0
  for j in range(4):
    a = np.sum(x[j]*w) + b_x*b
    output = 1/(1+np.exp(-a)) #sigmoid 함수 f(a)
    error = y[j][0] - output
    
    w= w + x[j] * 0.1 * error # 0.1:학습값?
    b= b + b_x * 0.1 * error
    
    error_sum = error_sum + error

  if i % 200 == 199: # 총 2000번 돌리는데 에러 확인을 위해 200번 마다 error print (나머지=199)
    print(i, error_sum, w.numpy(), b.numpy())
```

```
#확인용
for i  in range(4):
  a = np.sum(x[i]*w)+b_x*b
  print('X:', x[i], 'Y:', y[i], 'predicted:', 1/(1+np.exp(-a)))
```

```
#해결법
# Dense : 노드들을 빽빽하게 연결시켜줌, units : 연결되는 노드, activation : f함수 
# lr : learning_rate 바꿔가면서 좋은 결과 도출, loss : (y_hat - y)^2 = error
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(units=2, activation='sigmoid', input_shape = (2,)),
                             tf.keras.layers.Dense(units=1, activation='sigmoid')
])
model.compile(optimizer = tf.keras.optimizers.SGD(lr=0.3), loss = 'mse')

model.summary()
```

# XOR neural network
### 편향을 포함한 2-layer XOR 네트워크의 구조

- b=1, default로 존재 (입력1)
- b 사이에는 가중치(weight)존재x

```
# tf.keras를 이용한 XOR GATE 학습
history = model.fit(x, y, epochs=2000, batch_size=1) # epochs : 학습 횟수, batch_size : 학습할 때의 단위 (한번에 학습 할 데이터의 개수)
```
